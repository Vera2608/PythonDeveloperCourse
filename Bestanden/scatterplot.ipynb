```python
import pandas as pd
df = pd.read_csv("heartdisease.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_df = df.select_dtypes(include="number").dropna()
sns.set(style="whitegrid", context="notebook")
g = sns.pairplot(
    num_df,
    diag_kind="hist",      # or "kde"
    corner=False,          # True shows only lower triangle
    plot_kws={"alpha": 0.6, "s": 35}
)

g.fig.suptitle("Scatterplot Matrix", y=1.02)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("heartdisease.csv")
# df = pd.read_excel("dataset.xlsx")  # if Excel

# Optional: drop missing values in needed columns
plot_df = df[["sex", "age"]].dropna()
plot_df["sex"] = plot_df["sex"].map({0: "Female", 1: "Male"})

# Boxplot: sex vs chol
plt.figure(figsize=(7, 5))
sns.boxplot(data=plot_df, x="sex", y="age")
plt.title("Boxplot of Age by Sex")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.tight_layout()
plt.show()
```
