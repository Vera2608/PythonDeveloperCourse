```python
class Auto:
    def __init__(self, merk, model): #zo'n functie definieert data
        self.merk = merk
        self.model = model
        self.snelheid = 0
    def versnellen (self, delta): #zo'n functie kan logica bevatten / iets doen
        self.snelheid += delta
```

```python
a = Auto("Toyota", "Yaris")
b = Auto("Volvo", "ID4")

a.versnellen(25)
b.merk, a.snelheid, b.snelheid
```
